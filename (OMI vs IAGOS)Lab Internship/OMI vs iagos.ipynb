```python
import os
import re
import glob
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime, timedelta
from matplotlib.lines import Line2D  # Used for custom legend

# =======================
# Parameter Configuration
# =======================

# Root directory where data files are located
data_root_dir = r"/content/drive/MyDrive/(OMI vs IAGOS)Lab Internship/Data"

# OMI file matching pattern (corrected regular expression, added missing underscore)
omi_pattern = re.compile(r'^omi_l2_v2\+soft\+cmc_\d{4}m\d{4}_v\d{4}_BY\d_S\d+N\d+\.nc$', re.IGNORECASE)

# IAGOS file matching pattern (corrected regular expression, matching 16-digit timestamp)
iagos_pattern = re.compile(r'^IAGOS_profile_\d{16}_L2_.*\.nc4$', re.IGNORECASE)

# Output directory for saving images
output_dir = os.path.join("/content/drive/MyDrive/(OMI vs IAGOS)Lab Internship/Output", 'plots_combined')
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Output directory for saving CSVs
csv_output_dir = os.path.join(r"/content/drive/MyDrive/(OMI vs IAGOS)Lab Internship/Output/output_csv")
if not os.path.exists(csv_output_dir):
    os.makedirs(csv_output_dir)

# Latitude and longitude of ICN airport
icn_lat = 37.46
icn_lon = 126.44

# Earth radius (km)
earth_radius_km = 6371.0

# =======================
# Function Definitions
# =======================

def extract_timestamp(filename):
    """
    Extract timestamp from the filename.
    OMI filename example: omi_l2_v2+soft+cmc_2023m0502_v0401_BY2_S05N65.nc
    IAGOS filename example: IAGOS_profile_2023050203041714_L2_3.1.0-descent.nc4
    """
    if 'IAGOS' in filename:
        # Extract timestamp from IAGOS file, assuming format IAGOS_profile_YYYYMMDDHHMMSSXXXX_L2_...
        match = re.search(r'IAGOS_profile_(\d{16})', filename)
        if match:
            try:
                # Assume first 14 digits are YYYYMMDDHHMMSS, ignore last 2 digits
                timestamp_str = match.group(1)
                # Parse first 14 digits as datetime, ignore last 2 digits
                timestamp = datetime.strptime(timestamp_str[:14], '%Y%m%d%H%M%S')
                return timestamp
            except ValueError:
                print(f"Unable to parse timestamp in IAGOS file: {match.group(1)}")
                return None
    elif 'omi' in filename:
        # Extract date from OMI file, assuming format omi_l2_v2+soft+cmc_YYYYmMMDD_...
        match = re.search(r'(\d{4})m(\d{4})', filename)
        if match:
            year, month_day = match.groups()
            month = month_day[:2]
            day = month_day[2:]
            date_str = f"{year}{month}{day}"
            try:
                # Assume OMI file time is at 04:30 of that date
                timestamp = datetime.strptime(date_str, '%Y%m%d') + timedelta(hours=4, minutes=30)
                return timestamp
            except ValueError:
                print(f"Unable to parse date in OMI file: {date_str}")
                return None
    return None

def find_closest_iagos_file(omi_time, iagos_files):
    """
    Find the IAGOS file closest in time to the given OMI time.
    """
    min_diff = float('inf')
    best_match = None
    for f in iagos_files:
        iagos_time = extract_timestamp(f)
        if iagos_time:
            time_diff = abs((omi_time - iagos_time).total_seconds())
            if time_diff < min_diff:
                min_diff = time_diff
                best_match = f
    return best_match

def process_iagos_data(iagos_file_path, pressure_boundaries_omi, omi_height_layers):
    """
    Process IAGOS data, calculate average PPB and corresponding DU within each OMI pressure layer.
    """
    try:
        ds_iagos = xr.open_dataset(iagos_file_path)
    except Exception as e:
        print(f"      Error: Unable to open IAGOS file {iagos_file_path}: {e}")
        return None, None, None

    # Extract PPB and pressure data from IAGOS
    try:
        ppb_data = ds_iagos['O3_P1'].values  # Ozone concentration (ppb)
        pressure_data = ds_iagos['air_press_AC'].values / 100  # Convert pressure from Pa to hPa
    except KeyError as e:
        print(f"      Error: Missing necessary variable {e} in IAGOS file {iagos_file_path}.")
        return None, None, None

    # Read 'O3_P1_validity_flag' and filter
    try:
        ozone_flag = ds_iagos['O3_P1_validity_flag'].values  # Validity flag for ozone concentration
    except KeyError as e:
        print(f"      Error: Missing necessary variable {e} in IAGOS file {iagos_file_path}.")
        return None, None, None

    # Filter out points where validity flag is not 0
    valid_idx = ozone_flag == 0  # Keep only valid points

    # Apply filter to ppb_data and pressure_data
    ppb_data = ppb_data[valid_idx]
    pressure_data = pressure_data[valid_idx]

    print(f"      Processing IAGOS file: {os.path.basename(iagos_file_path)}")
    print(f"        Original PPB data length: {len(ppb_data)}, Original pressure data length: {len(pressure_data)}")

    # Save original IAGOS data to CSV
    iagos_filename = os.path.basename(iagos_file_path)
    iagos_timestamp = extract_timestamp(iagos_filename)
    iagos_datetime_str = iagos_timestamp.strftime("%Y%m%d%H%M%S") if iagos_timestamp else 'unknown'

    df_iagos_original = pd.DataFrame({
        'Ozone_Concentration_ppb': ppb_data,
        'Pressure_hPa': pressure_data,
    })
    iagos_original_csv_filename = f'IAGOS_Original_Data_{iagos_datetime_str}.csv'
    iagos_original_csv_path = os.path.join(csv_output_dir, iagos_original_csv_filename)
    df_iagos_original.to_csv(iagos_original_csv_path, index=False)
    print(f"      Original IAGOS data saved to {iagos_original_csv_path}")

    # Remove NaN values
    mask_valid = ~np.isnan(ppb_data) & ~np.isnan(pressure_data)
    ppb_values = ppb_data[mask_valid]
    pressure_values = pressure_data[mask_valid]

    print(f"        After removing NaN, PPB data length: {len(ppb_values)}, Pressure data length: {len(pressure_values)}")
    print(f"        IAGOS pressure range: {pressure_values.min()} - {pressure_values.max()} hPa")

    # Save valid IAGOS data to CSV
    df_iagos_valid = pd.DataFrame({
        'Ozone_Concentration_ppb': ppb_values,
        'Pressure_hPa': pressure_values,
    })
    iagos_valid_csv_filename = f'IAGOS_Valid_Data_{iagos_datetime_str}.csv'
    iagos_valid_csv_path = os.path.join(csv_output_dir, iagos_valid_csv_filename)
    df_iagos_valid.to_csv(iagos_valid_csv_path, index=False)
    print(f"      Valid IAGOS data saved to {iagos_valid_csv_path}")

    # Check if data points are sufficient
    if len(ppb_values) < 2:
        print("        Error: Insufficient IAGOS data points, cannot process.")
        return None, None, None

    # Sort data by pressure in descending order (from high pressure to low pressure)
    sorted_indices = np.argsort(-pressure_values)
    ppb_values = ppb_values[sorted_indices]
    pressure_values = pressure_values[sorted_indices]

    # Calculate average PPB and DU within each OMI pressure layer
    avg_ppb_iagos = []
    du_iagos = []
    avg_pressure_iagos = []

    # Get minimum pressure value from original IAGOS data
    min_pressure_iagos = pressure_values.min()

    for i in range(len(pressure_boundaries_omi) - 1):
        P_top = pressure_boundaries_omi[i]
        P_bottom = pressure_boundaries_omi[i + 1]

        # If the bottom pressure of the current layer is less than the minimum pressure of IAGOS data, skip this layer
        if P_bottom < min_pressure_iagos:
            print(f"        Layer {i}: Bottom pressure {P_bottom:.2f} hPa is less than IAGOS minimum pressure {min_pressure_iagos:.2f} hPa, skipping this layer.")
            avg_ppb_iagos.append(np.nan)
            du_iagos.append(np.nan)
            avg_pressure_iagos.append((P_top + P_bottom) / 2)
            continue

        # Find indices of IAGOS data points within the current OMI pressure layer
        mask = (pressure_values <= P_top) & (pressure_values > P_bottom)
        if np.any(mask):
            ppb_in_layer = ppb_values[mask]
            pressure_in_layer = pressure_values[mask]

            # Calculate ΔP_i
            delta_p_in_layer = np.diff(pressure_in_layer, prepend=pressure_in_layer[0])

            # Calculate weighted average PPB
            weighted_ppb = np.sum(ppb_in_layer * delta_p_in_layer) / np.sum(delta_p_in_layer)
            avg_ppb_iagos.append(weighted_ppb)

            # Calculate pressure difference ΔP (hPa)
            delta_p = P_top - P_bottom

            # Calculate z_mid (km)
            z_mid = omi_height_layers[i]

            # Calculate DU
            du = (weighted_ppb * delta_p) / (1.251 * 1013.25) * ((earth_radius_km + z_mid) / earth_radius_km)**2
            du_iagos.append(du)

            # Calculate average pressure (for plotting)
            avg_pressure = (P_top + P_bottom) / 2
            avg_pressure_iagos.append(avg_pressure)
        else:
            avg_ppb_iagos.append(np.nan)
            du_iagos.append(np.nan)
            avg_pressure_iagos.append((P_top + P_bottom) / 2)

    avg_ppb_iagos = np.array(avg_ppb_iagos)
    du_iagos = np.array(du_iagos)
    avg_pressure_iagos = np.array(avg_pressure_iagos)

    print("        Resampled IAGOS average PPB values (before DU calculation):")
    print(avg_ppb_iagos)
    print("        Resampled IAGOS DU values:")
    print(du_iagos)
    print("        Resampled IAGOS average pressure values:")
    print(avg_pressure_iagos)

    # Check array lengths
    print("Lengths of arrays before creating DataFrame:")
    print(f"Length of avg_ppb_iagos: {len(avg_ppb_iagos)}")
    print(f"Length of du_iagos: {len(du_iagos)}")
    print(f"Length of avg_pressure_iagos: {len(avg_pressure_iagos)}")

    # Save resampled IAGOS data to CSV
    df_iagos_resampled = pd.DataFrame({
        'Layer_Index': np.arange(len(avg_ppb_iagos)),
        'Pressure_Boundary_Top_hPa': pressure_boundaries_omi[:-1],
        'Pressure_Boundary_Bottom_hPa': pressure_boundaries_omi[1:],
        'Average_Pressure_hPa': avg_pressure_iagos,
        'Average_Ozone_Concentration_ppb': avg_ppb_iagos,
        'DU': du_iagos,
    })
    iagos_resampled_csv_filename = f'IAGOS_Resampled_Data_{iagos_datetime_str}.csv'
    iagos_resampled_csv_path = os.path.join(csv_output_dir, iagos_resampled_csv_filename)
    df_iagos_resampled.to_csv(iagos_resampled_csv_path, index=False)
    print(f"      Resampled IAGOS data saved to {iagos_resampled_csv_path}")

    return avg_ppb_iagos, du_iagos, avg_pressure_iagos

def plot_profiles(omi_data, iagos_data, output_path, omi_date_str, iagos_filename):
    """
    Plot O3 profiles from OMI and IAGOS over ICN Airport, and save to the specified path.
    """
    (icn_ozprof_omi, icn_aozprof_omi, average_ps_hpa_omi, average_zs_km_omi,
     icn_rms_omi, icn_cfrac_omi, icn_avgres_omi, pressure_boundaries_omi) = omi_data
    (avg_ppb_iagos, du_iagos, avg_pressure_iagos) = iagos_data

    # Take log10 of average pressure for plotting
    log_average_ps_omi = np.log10(average_ps_hpa_omi)

    # Only keep the lowest 4 data points
    du_iagos_plot = du_iagos[:4]
    avg_pressure_iagos_plot = avg_pressure_iagos[:4]
    log_avg_pressure_iagos_plot = np.log10(avg_pressure_iagos_plot)

    # Print resampled IAGOS data
    print("        Resampled IAGOS data (lowest 4 data points):")
    for idx in range(len(du_iagos_plot)):
        print(f"          DU[{idx+1}]: {du_iagos_plot[idx]:.4f}, Pressure[{idx+1}]: {avg_pressure_iagos_plot[idx]:.2f} hPa")

    # Create figure
    fig, ax1 = plt.subplots(figsize=(10, 8))  # Increase figure size for better readability

    # Plot OMI O3 profile
    ax1.plot(icn_ozprof_omi, log_average_ps_omi, marker='o', linestyle='-', color='b', label='OMI O3 Concentration (ozprof)')

    # Plot OMI a priori O3 profile
    ax1.plot(icn_aozprof_omi, log_average_ps_omi, marker='x', linestyle='--', color='r', label='OMI A Priori O3 Concentration (aozprof)')

    # Plot the lowest 4 DU data points from IAGOS
    if np.any(~np.isnan(du_iagos_plot)):
        ax1.plot(du_iagos_plot, log_avg_pressure_iagos_plot, marker='s', linestyle='--', color='g', label='IAGOS O3 Concentration (DU)')
    else:
        print("        Warning: All of the lowest 4 IAGOS data points are NaN, not plotting this curve.")

    # Define RMS and Cloud Fraction information
    rms_text_win1 = f"RMS (win1): {icn_rms_omi[0]:.2f}"
    rms_text_win2 = f"RMS (win2): {icn_rms_omi[1]:.2f}"
    cfrac_text = f"Cloud Fraction: {icn_cfrac_omi:.2f}"

    # Define Avg Res information
    avgres_text_win1 = f"Avg Res (win1): {icn_avgres_omi[0]:.2f}"
    avgres_text_win2 = f"Avg Res (win2): {icn_avgres_omi[1]:.2f}"

    # Create custom legend elements, including additional text information
    legend_elements = [
        Line2D([0], [0], color='b', marker='o', linestyle='-', label='OMI O3 Concentration (ozprof)'),
        Line2D([0], [0], color='r', marker='x', linestyle='--', label='OMI A Priori O3 Concentration (aozprof)'),
    ]

    if np.any(~np.isnan(du_iagos_plot)):
        legend_elements.append(Line2D([0], [0], color='g', marker='s', linestyle='--', label='IAGOS O3 Concentration (DU)'))

    # Add RMS, Cloud Fraction, Avg Res information
    legend_elements.extend([
        Line2D([0], [0], color='none', label=rms_text_win1),
        Line2D([0], [0], color='none', label=rms_text_win2),
        Line2D([0], [0], color='none', label=cfrac_text),
        Line2D([0], [0], color='none', label=avgres_text_win1),
        Line2D([0], [0], color='none', label=avgres_text_win2),
    ])

    # Add legend
    ax1.legend(handles=legend_elements, loc='upper right', fontsize=10)

    # Set title and axis labels
    ax1.set_title(f'O3 Vertical Profile over ICN Airport (OMI & IAGOS) - {omi_date_str}', fontsize=14)
    ax1.set_xlabel('O3 (DU)', fontsize=12)

    # Set pressure axis ticks
    pressure_ticks_hpa = [1000, 100, 10, 1, 0.1]
    log_pressure_ticks = np.log10(pressure_ticks_hpa)
    ax1.set_yticks(log_pressure_ticks)
    ax1.set_yticklabels([f"{p}" for p in pressure_ticks_hpa])
    ax1.set_ylabel('Pressure (hPa)', fontsize=12)

    # Create secondary y-axis for altitude
    ax2 = ax1.twinx()

    # Set altitude axis limits based on height data
    min_height = average_zs_km_omi.min()
    max_height = average_zs_km_omi.max()
    ax2.set_ylim(min_height, max_height)

    # Set fixed altitude ticks
    fixed_height_ticks = [0, 10, 20, 30, 40, 50, 60, 70]
    ax2.set_yticks(fixed_height_ticks)
    ax2.set_yticklabels([f"{h}" for h in fixed_height_ticks])
    ax2.set_ylabel('Height (km)', fontsize=12)

    # Add grid
    ax1.grid(True, which='both', linestyle='--', linewidth=0.5)

    # Invert pressure axis so that high pressure is at the bottom
    ax1.invert_yaxis()
    # Do not invert altitude axis

    # Adjust layout to prevent labels from being cut off
    plt.tight_layout()

    # Save figure
    plt.savefig(output_path, dpi=300)
    plt.show()
    plt.close()

def main():
    """
    Main program: iterate over all OMI files, find matching IAGOS files, process data, and generate plots.
    """
    # Use glob to initially filter files
    omi_glob_pattern = os.path.join(data_root_dir, 'omi_l2_v2+soft+cmc_*.nc')
    iagos_glob_pattern = os.path.join(data_root_dir, 'IAGOS_profile_*_L2_*.nc4')

    omi_glob_files = glob.glob(omi_glob_pattern)
    iagos_glob_files = glob.glob(iagos_glob_pattern)

    # Extract filenames and match with regex
    omi_files = [os.path.basename(f) for f in omi_glob_files if omi_pattern.match(os.path.basename(f))]
    iagos_files = [os.path.basename(f) for f in iagos_glob_files if iagos_pattern.match(os.path.basename(f))]

    print(f"\nFound {len(omi_files)} OMI files and {len(iagos_files)} IAGOS files.")
    print("\nOMI files:")
    for f in omi_files:
        print(f"  {f}")
    print("\nIAGOS files:")
    for f in iagos_files:
        print(f"  {f}")

    # Exit if no OMI or IAGOS files found
    if not omi_files:
        print("No OMI files found. Please check the file naming pattern.")
        return
    if not iagos_files:
        print("No IAGOS files found. Please check the file naming pattern.")
        return

    # Iterate over each OMI file
    for omi_file in omi_files:
        omi_file_path = os.path.join(data_root_dir, omi_file)
        omi_datetime = extract_timestamp(omi_file)

        if not omi_datetime:
            print(f"  Skipping OMI file (unable to extract timestamp): {omi_file}")
            continue

        # Print current OMI file being processed
        print(f"\nProcessing OMI file: {omi_file} (Timestamp: {omi_datetime})")

        try:
            # Open OMI NetCDF file using xarray
            ds_omi = xr.open_dataset(omi_file_path)

            # Rename duplicate dimension name, only rename 'nl' to 'nl1'
            ds_omi = ds_omi.rename_dims({'nl': 'nl1'})
        except Exception as e:
            print(f"  Error: Unable to open OMI file {omi_file}: {e}")
            continue

        # Extract variables
        try:
            ozprof_omi = ds_omi['ozprof'].values      # Vertical ozone profile (nl1, mpix)
            aozprof_omi = ds_omi['aozprof'].values    # A priori ozone profile (nl1, mpix)
            ps_omi = ds_omi['ps'].values              # Pressure boundaries (nlp1, mpix)
            zs_omi = ds_omi['zs'].values              # Altitude boundaries (nlp1, mpix)
            rms_omi = ds_omi['rms'].values            # Root mean square error (win, mpix)
            cfrac_omi = ds_omi['cfrac'].values        # Cloud fraction (mpix)
            avgres_omi = ds_omi['avgres'].values      # Average residual (win, mpix)
        except KeyError as e:
            print(f"  Error: Missing necessary variable {e} in OMI file {omi_file}.")
            continue

        # Extract latitude and longitude
        try:
            lat_omi = ds_omi['lat'].values
            lon_omi = ds_omi['lon'].values
        except KeyError as e:
            print(f"  Error: Missing latitude or longitude variable {e} in OMI file {omi_file}.")
            continue

        # Find the data point closest to ICN
        distance_omi = np.sqrt((lat_omi - icn_lat)**2 + (lon_omi - icn_lon)**2)
        nearest_index_omi = distance_omi.argmin()

        nearest_lat_omi = lat_omi[nearest_index_omi]
        nearest_lon_omi = lon_omi[nearest_index_omi]

        print(f"  Selected nearest coordinates in OMI: Latitude = {nearest_lat_omi}, Longitude = {nearest_lon_omi}")

        # Get O3 vertical profile and pressure/altitude information at ICN
        icn_ozprof_omi = ozprof_omi[:, nearest_index_omi]    # Ozone concentration at each layer (nl1)
        icn_aozprof_omi = aozprof_omi[:, nearest_index_omi]  # A priori ozone concentration at each layer (nl1)
        icn_ps_omi = ps_omi[:, nearest_index_omi]            # Pressure boundaries (nlp1)
        icn_zs_omi = zs_omi[:, nearest_index_omi]            # Altitude boundaries (nlp1)
        icn_rms_omi = rms_omi[:, nearest_index_omi]          # RMS data (win, )
        icn_cfrac_omi = cfrac_omi[nearest_index_omi]         # Cloud fraction
        icn_avgres_omi = avgres_omi[:, nearest_index_omi]    # Average residual data (win, )

        # Reverse pressure and altitude boundary arrays so that they are ordered from high pressure to low pressure
        icn_ps_omi = icn_ps_omi[::-1]
        icn_zs_omi = icn_zs_omi[::-1]

        # Pressure boundary data
        pressure_boundaries_omi = icn_ps_omi.copy()  # Pressure boundaries (hPa)
        height_boundaries_omi = icn_zs_omi.copy()    # Altitude boundaries (km)

        # Calculate average pressure and altitude for each layer
        average_ps_hpa_omi = (pressure_boundaries_omi[:-1] + pressure_boundaries_omi[1:]) / 2  # Average pressure per layer (hPa)
        average_zs_km_omi = (height_boundaries_omi[:-1] + height_boundaries_omi[1:]) / 2       # Average altitude per layer (km)

        # Also reverse OMI ozone concentration data arrays
        icn_ozprof_omi = icn_ozprof_omi[::-1]
        icn_aozprof_omi = icn_aozprof_omi[::-1]

        print(f"  OMI pressure boundaries (descending): {pressure_boundaries_omi}")
        print(f"  OMI average pressure per layer (descending): {average_ps_hpa_omi}")
        print(f"  OMI average altitude per layer (corresponding to descending pressure): {average_zs_km_omi}")

        # Save OMI data to CSV
        # Create a DataFrame containing OMI data
        omi_datetime_str = omi_datetime.strftime("%Y%m%d%H%M%S")
        df_omi = pd.DataFrame({
            'Layer_Index': np.arange(len(icn_ozprof_omi)),
            'Pressure_Boundary_Top_hPa': pressure_boundaries_omi[:-1],
            'Pressure_Boundary_Bottom_hPa': pressure_boundaries_omi[1:],
            'Average_Pressure_hPa': average_ps_hpa_omi,
            'Average_Height_km': average_zs_km_omi,
            'OMI_Ozone_Concentration_ozprof': icn_ozprof_omi,
            'OMI_A_Priori_Ozone_Concentration_aozprof': icn_aozprof_omi,
        })

        # Add RMS, Cloud Fraction, AvgRes data
        df_omi['RMS_win1'] = icn_rms_omi[0] if len(icn_rms_omi) > 0 else np.nan
        df_omi['RMS_win2'] = icn_rms_omi[1] if len(icn_rms_omi) > 1 else np.nan
        df_omi['Cloud_Fraction'] = icn_cfrac_omi
        df_omi['AvgRes_win1'] = icn_avgres_omi[0] if len(icn_avgres_omi) > 0 else np.nan
        df_omi['AvgRes_win2'] = icn_avgres_omi[1] if len(icn_avgres_omi) > 1 else np.nan

        # Generate filename
        omi_csv_filename = f'OMI_ICN_Data_{omi_datetime_str}.csv'
        omi_csv_path = os.path.join(csv_output_dir, omi_csv_filename)

        # Save to CSV
        df_omi.to_csv(omi_csv_path, index=False)
        print(f"  OMI data saved to {omi_csv_path}")

        # Find the IAGOS file closest in time to the current OMI file
        closest_iagos_file = find_closest_iagos_file(omi_datetime, iagos_files)

        if not closest_iagos_file:
            print(f"  No IAGOS file found close in time to the OMI file.")
            continue

        iagos_file_path = os.path.join(data_root_dir, closest_iagos_file)
        print(f"  Matched IAGOS file: {closest_iagos_file}")

        # Process IAGOS data
        avg_ppb_iagos, du_iagos, avg_pressure_iagos = process_iagos_data(
            iagos_file_path, pressure_boundaries_omi, average_zs_km_omi
        )

        if avg_ppb_iagos is None or du_iagos is None:
            print(f"  Skipping OMI file (failed to process IAGOS data): {omi_file}")
            continue

        # Prepare data needed for plotting
        omi_data = (
            icn_ozprof_omi,
            icn_aozprof_omi,
            average_ps_hpa_omi,
            average_zs_km_omi,
            icn_rms_omi,  # Keep all win data
            icn_cfrac_omi,
            icn_avgres_omi,
            pressure_boundaries_omi
        )
        iagos_data = (avg_ppb_iagos, du_iagos, avg_pressure_iagos)

        # Ensure array lengths are consistent
        min_length = min(
            len(icn_ozprof_omi),
            len(icn_aozprof_omi),
            len(average_ps_hpa_omi),
            len(average_zs_km_omi),
            len(avg_ppb_iagos),
            len(du_iagos),
            len(avg_pressure_iagos),
        )

        # Truncate arrays
        layer_index = np.arange(min_length)
        average_ps_hpa_omi = average_ps_hpa_omi[:min_length]
        average_zs_km_omi = average_zs_km_omi[:min_length]
        icn_ozprof_omi = icn_ozprof_omi[:min_length]
        icn_aozprof_omi = icn_aozprof_omi[:min_length]
        avg_ppb_iagos = avg_ppb_iagos[:min_length]
        du_iagos = du_iagos[:min_length]
        avg_pressure_iagos = avg_pressure_iagos[:min_length]

        # Save plotting data to CSV
        iagos_timestamp = extract_timestamp(closest_iagos_file)
        iagos_datetime_str = iagos_timestamp.strftime("%Y%m%d%H%M%S") if iagos_timestamp else 'unknown'

        df_plot_data = pd.DataFrame({
            'Layer_Index': layer_index,
            'Average_Pressure_hPa': average_ps_hpa_omi,
            'Average_Height_km': average_zs_km_omi,
            'OMI_Ozone_Concentration_ozprof': icn_ozprof_omi,
            'OMI_A_Priori_Ozone_Concentration_aozprof': icn_aozprof_omi,
            'IAGOS_Average_Ozone_Concentration_ppb': avg_ppb_iagos,
            'IAGOS_DU': du_iagos,
        })

        # Generate filename
        plot_data_csv_filename = f'Plot_Data_{omi_datetime_str}_IAGOS_{iagos_datetime_str}.csv'
        plot_data_csv_path = os.path.join(csv_output_dir, plot_data_csv_filename)

        df_plot_data.to_csv(plot_data_csv_path, index=False)
        print(f"  Plot data saved to {plot_data_csv_path}")

        # Generate output image filename
        output_filename = f'O3_Profile_ICN_{omi_datetime.strftime("%Y%m%d%H%M%S")}_IAGOS_{closest_iagos_file}.png'
        output_path = os.path.join(output_dir, output_filename)

        # Plot and save the figure
        plot_profiles(omi_data, iagos_data, output_path, omi_datetime.strftime('%Y-%m-%d %H:%M:%S'), closest_iagos_file)
        print(f"  Plot saved to {output_path}")

if __name__ == "__main__":
    main()
```
